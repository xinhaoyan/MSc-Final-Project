```python
with open('output.txt', 'w', encoding='utf-8') as file:
    for index, row in data.iterrows():
        if row["new_x"] != row["old_x"]:
            file.write(f"用户关注从{row['old_x']}内容变化到{row['new_x']}\n")
        if row["new_y"] != row["old_y"]:
            file.write(f"用户关注从{row['old_y']}内容变化到{row['new_y']}\n")
        if row["new_size"] != row["old_size"]:
            file.write(f"用户关注从{row['old_size']}内容变化到{row['new_size']}\n")
        if row["new_color"] != row["old_color"]:
            file.write(f"用户关注从{row['old_color']}内容变化到{row['new_color']}\n")
        if row["new_year"] != row["old_year"]:
            file.write(f"Users focus time and have switched from {row['old_year']} year to {row['new_year']} year \n")
        if row["new_country"] != row["old_country"]:
            # 注意这里我们处理了row['new_country']，而不是整个data['new_country']
            countries_list = row['new_country'].strip("[]").split(";")
            countries_str = "; ".join([f"'{country.strip()}'" for country in countries_list])
            file.write(f"User focus country and choose {countries_str} country\n")
print("All rows processed")
```
